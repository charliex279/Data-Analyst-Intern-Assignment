```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/Data.xlsx'

user_details = pd.read_excel(file_path)
cooking_sessions = pd.read_excel(file_path, sheet_name=1)
order_details = pd.read_excel(file_path, sheet_name=2)
```

```python
# Handling missing values

print(user_details.isnull().sum())
print(cooking_sessions.isnull().sum())
print(order_details.isnull().sum())
```

```python
def clean_data(df):

  # remove duplicates. Duplicates in the data can distort analyses, such as inflating counts or skewing averages

    df = df.drop_duplicates()

  #Handling missing values. Missing values (e.g., NaN) in the dataset can disrupt calculations, visualizations, and machine learning algorithms

    df = df.dropna()

    return df

user_details = clean_data(user_details)
cooking_sessions = clean_data(cooking_sessions)
order_details = clean_data(order_details)
```

```python
display(order_details)
```

```python
#  Data Merging using the common column values

merged_data = pd.merge(user_details, cooking_sessions, on='User ID', how='inner')
merged_data = pd.merge(merged_data, order_details, on=['User ID', 'Session ID'], how='inner')
```

```python
print(merged_data.columns)
```

```python
# Group Users and Aggregate Data using  groupby('User ID') groups the data by each unique user.

cooking_order_relation = merged_data.groupby('User ID').agg({
    'Session ID': 'nunique',
    'Order ID': 'nunique'
}).rename(columns={'Session ID': 'CookingSessions', 'Order ID': 'Orders'})

#Correlation Analysis:(.corr()) method calculates the correlation coefficient between CookingSessions and Orders.

#Correlation measures the strength and direction of the relationship between two variables.

correlation = cooking_order_relation.corr()
print("Correlation between cooking sessions and orders:")
print(correlation)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=cooking_order_relation, x='CookingSessions', y='Orders')
plt.title('Relationship between Cooking Sessions and Orders')
plt.xlabel('Number of Cooking Sessions')
plt.ylabel('Number of Orders')

x_values = cooking_order_relation['CookingSessions']
slope = 1
intercept = 0
y_values = slope * x_values + intercept

plt.plot(x_values, y_values, color='red', linestyle='--', label='Dividing Line')

plt.legend()
plt.show()
```

```python
# Age Distribution of Users

plt.figure(figsize=(8, 6))
sns.histplot(user_details['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Identifying popular dishes

popular_dishes = merged_data['Dish Name_y'].value_counts().head(10)
print("Top 10 popular dishes:")
print(popular_dishes)

plt.figure(figsize=(10, 6))
popular_dishes.plot(kind='bar', color='skyblue')
plt.title('Top 10 Popular Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
# The groupby('Age') part groups the data by each unique age value, and .agg() computes the averages for each group

# analyzes how user age influences the duration of sessions and spending behavior.

demographics_analysis = merged_data.groupby('Age').agg({
    'Duration (mins)': 'mean',
    'Amount (USD)': 'mean'
}).reset_index()

# provides demographic trends regarding session duration and spending behavior.

print("Demographic factors influencing behavior:")
print(demographics_analysis)

plt.figure(figsize=(8, 6))
sns.lineplot(data=demographics_analysis, x='Age', y='Duration (mins)', label='Average Duration (mins)', marker='o')
sns.lineplot(data=demographics_analysis, x='Age', y='Amount (USD)', label='Average Amount (USD)', marker='o')
plt.title('Demographic Analysis of User Behavior')
plt.xlabel('Age')
plt.ylabel('Average Values')
plt.legend()
plt.show()
```

```python
# Favorite Meal Popularity

favorite_meals = user_details['Favorite Meal'].value_counts().head(10)
plt.figure(figsize=(10, 6))
favorite_meals.plot(kind='bar', color='purple')
plt.title('Top 10 Favorite Meals of Users')
plt.xlabel('Favorite Meal')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()
```

```python
#Total Orders by Age Group

orders_by_age = user_details.groupby('Age').agg({'Total Orders': 'sum'}).reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=orders_by_age, x='Age', y='Total Orders', marker='o', color='green')
plt.title('Total Orders by Age Group')
plt.xlabel('Age')
plt.ylabel('Total Orders')
plt.show()
```

```python
#Sessions Ratings

plt.figure(figsize=(8, 6))
sns.boxplot(data=cooking_sessions, x='Session Rating', color='lightblue')
plt.title('Session Ratings Distribution')
plt.xlabel('Session Rating')
plt.show()
```

```python
# Total Orders by Location, Grouped by Age and Meal Type

location_analysis = merged_data.groupby(['Location', 'Age', 'Meal Type_y']).agg({
    'Order ID': 'nunique',
    'Amount (USD)': 'sum'
}).reset_index()

plt.figure(figsize=(12, 8))

sns.barplot(data=location_analysis,
            x='Location',
            y='Order ID',
            hue='Meal Type_y',
            ci=None)

plt.title('Total Orders by Location, Grouped by Age and Meal Type')
plt.xlabel('Location')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.legend(title='Meal Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside
plt.tight_layout()


plt.show()
```

```python
#Revenue by Time of Day

revenue_by_time = merged_data.groupby('Time of Day').agg({'Amount (USD)': 'sum'}).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=revenue_by_time, x='Time of Day', y='Amount (USD)', palette='Blues')
plt.title('Revenue by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Revenue (USD)')
plt.show()
```

```python
user_details['Registration Date'] = pd.to_datetime(user_details['Registration Date'])
orders_by_registration = user_details.groupby(user_details['Registration Date'].dt.to_period('M')).agg({
    'Total Orders': 'sum'
}).reset_index()


orders_by_registration['Registration Date'] = orders_by_registration['Registration Date'].dt.to_timestamp()

plt.figure(figsize=(10, 6))
sns.lineplot(data=orders_by_registration, x='Registration Date', y='Total Orders', marker='o', color='brown')
plt.title('Total Orders by Registration Date')
plt.xlabel('Registration Month')
plt.ylabel('Total Orders')
plt.show()
```

```python
# Count occurrences of each meal type

meal_type_counts = merged_data['Meal Type_y'].value_counts()

plt.figure(figsize=(8, 6))
meal_type_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('pastel'),
    textprops={'fontsize': 12}
)
plt.title('Meal Type Distribution')
plt.ylabel('')
plt.show()
```
